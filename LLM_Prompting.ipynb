```python
from openai import OpenAI
import requests
from bs4 import BeautifulSoup
import json

# API Setup
api_key = "ddc-a4f-49e9ee762ff445bcb1146ffb238d75b3"
api_base_url = "https://api.a4f.co/v1"
client = OpenAI(api_key=api_key, base_url=api_base_url)

def ask_tutor(question):
    response = client.chat.completions.create(
        model="provider-2/gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are a helpful assistant that answers the asked question precisely. Please provide different examples"},
            {"role": "user", "content": question}
        ]
    )

    return response.choices[0].message.content

# Example usage
print(ask_tutor("Explain for loop in Python"))

```
